```python
import pandas as pd
import numpy as np
train = pd.read_json('train.json')
test = pd.read_json('test.json')
print(train.shape)
print(test.shape)
print('\ntraining sample \n', train.sample(10))
print('\ntest sample \n',test.sample(10))
```

Next, we explore the data to see if there are cuisines with less than 3 ingredients. This also helps to check whether there are entries with no ingredients. We realise that there are no null entries in the dataset :)

```python
z = train.copy()
z['ingredients'] = z['ingredients'].apply(lambda x: x if len(str(x).split())< 3 else np.nan)
z = z.dropna()
z
```

Next, we see the distribution in our target labels. We realise that the dataset is lot more skewed towards some type of cuisines.

```python
train['cuisine'].value_counts()
```

```python
train['all_ingredients'] = train['ingredients'].map(";".join)
```

```python
from textblob import Word
train['all_ingredients'] = train['all_ingredients'].apply(lambda x: " ".join(x.lower() for x in x.split()))
train['all_ingredients'] = train['all_ingredients'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X = cv.fit_transform(train['all_ingredients'].values)
```

```python
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
y = enc.fit_transform(train.cuisine)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression()
logistic.fit(X_train, y_train)
logistic.score(X_test, y_test)
```
